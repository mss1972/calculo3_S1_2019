```python
import numpy as np
import matplotlib.pyplot as mplot
from itertools import cycle
```

Creando en $x$ un vector de $\mathbb{R}^n$ con $n \in \mathbb{N}$.

Puedes cambiar este vector.

```python
x = np.array([2,5,1,0,-1,4,8])
```

Ciclo para graficar $\|x\|_p$ para distintos valores de $p$ y ver que converge a $\|x\|_{\infty}$.

$N$ es la cantidad de normas a calcular. También puedes cambiar este valor. Mientras mayor sea el valor de $N$, mejor se podrá comprobar que la norma $p$ de un vector converge a la norma infinito del vector.

```python
N = 30
normas_x = np.zeros(N)
```

La primera componente del vector normas_x es la norma 1 de $x$

```python
normas_x[0] = np.linalg.norm(x,1)
```

La última componente del vector normas_x es la norma infinito de $x$.

```python
if N > 1:
    normas_x[-1] = np.linalg.norm(x,np.inf)
```

Las restantes componentes son la norma $p$ con $p \in \mathbb{N}, p \ge 2$

```python
if N > 2:
    for p in range(2,N):
        normas_x[p-1] = np.linalg.norm(x,p)
mplot.plot(range(1,N),normas_x[:-1],'bo')
mplot.plot(N+1,normas_x[-1],'ro')          
```

Película que muestra los vectores en $\mathbb{R}^2$ con norma $p$ igual a 1, $p \in \mathbb{N}, p \ge 1$ y los vectores en $\mathbb{R}^2$ con norma infinito igual a 1.

Si $x \in \mathbb{R}^2$, entonces $x = (x_1, x_2)$.

Note que $\|x\|_p = 1 \Rightarrow \left|x_1\right| \le 1 \wedge \left|x_2\right| \le 1$.

Para calcular los vectores en $\mathbb{R}^2$ con norma $p$ igual a 1, se deja variar $x_1$ entre -1 y 1 y se calcula $x_2$ de modo que $|x_1|^p + |x_2|^p = 1$, es decir, dado un valor de $x_1$ entre -1 y 1, la norma $p$ de x es 1 para $x_2 = \left(1-\left|x_1\right|^p\right)^{1/p}$ o $x_2 = -\left(1-\left|x_1\right|^p\right)^{1/p}$.

```python
# numero de valores de p, p > 1
Np = 10
# cantidad de vectores con norma p igual a 1 que se van a generar es 2N
N = 100
x = np.linspace(-1,1,N)
yp = np.zeros_like(x)
yn = np.zeros_like(x)
# generating colors
cycol = cycle('bgrcmk')
```

```python
# calculando los vectores con norma 1 igual a 1
yp = 1-abs(x)
yn = -yp
color1 = next(cycol)
mplot.plot(x,yp,c=color1)
mplot.plot(x,yn,c=color1)
for p in range(2,Np+2):
    yp = (1-abs(x)**p)**(1/p)
    yn = -yp
    mplot.plot(x,yp,x,yn,c=next(cycol))
# vectores con norma infinito igual a 1
color = next(cycol)
mplot.plot(x,np.ones_like(x),x,-np.ones_like(x),c=color)
mplot.plot(np.ones_like(x),x,-np.ones_like(x),x,c=color)
xmin,xmax,ymin,ymax = mplot.axis('scaled')
```

El rombo en azul son los vectores en $\mathbb{R}^2$ con norma 1 igual a 1.
La circunferencia son los vectores en $\mathbb{R}^2$ con norma 2 igual a 1.
A medida que $p$ aumenta, el lugar geométrico de los puntos en $\mathbb{R}^2$ con norma $p$ igual a 1 se acerca al de los puntos en $\mathbb{R}^2$ con norma infinito igual a 1.
